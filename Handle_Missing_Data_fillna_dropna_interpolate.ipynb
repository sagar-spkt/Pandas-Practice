```python
df_date_parsed = pd.read_csv("weather_data.csv", parse_dates=['day', ])
df_date_parsed
```

```python
df_date_parsed.set_index('day', inplace=True)
df_date_parsed
```

```python
df_date_parsed.fillna(0)
```

```python
df_date_parsed.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event',
})
```

```python
df_date_parsed.fillna(method='ffill')  # carry forward previous row value
# df_date_parsed.fillna(method='bfill')  # carry backward next row value
```

```python
df_date_parsed.interpolate()
```

```python
df_date_parsed.interpolate(method='time')
```
