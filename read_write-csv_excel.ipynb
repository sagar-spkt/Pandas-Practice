Passing header=None means csv doesn't contain header

```python
import pandas as pd
df_read_csv = pd.read_csv("stock_data.csv", header=1)
# OR df_read_csv = pd.read_csv("stock_data.csv", skiprows=1)
df_read_csv
```

```python
pd.read_csv("stock_data.csv", skiprows=1, nrows=3)
```

```python
pd.read_csv("stock_data.csv", na_values=['not available', 'n.a.'], header=1)
```

```python
pd.read_csv("stock_data.csv", na_values=['not available', 'n.a.'], header=1)
```

```python
df_with_na_values = pd.read_csv("stock_data.csv", na_values={
    'tickers': ['not available', 'n.a.'],
    'eps': ['not available', 'n.a.'],
    'revenue': ['not available', 'n.a.', -1],  # revenue cannot be negative
    'price': ['not available', 'n.a.', -1],
    'people': ['not available', 'n.a.'],
}, header=1)
df_with_na_values
```

```python
df_with_na_values.to_csv('new_csv.csv', index=False)
```

```python
df_with_na_values.to_csv('new_csv_only_specified_rows.csv', columns=['tickers', 'eps'], index=False)
```
