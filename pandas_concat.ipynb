```python
import pandas as pd

nepal_weather = pd.DataFrame({
    "city": ["kathmandu", "pokhara", "dharan"],
    "temperature": [32, 45, 30],
    "humidity": [80, 60, 78]
})
nepal_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([nepal_weather, us_weather])
df
```

```python
df = pd.concat([nepal_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([nepal_weather, us_weather], keys=['Nepal', 'US'])
df
```

```python
df.loc['Nepal']
```

```python
temperature_df = pd.DataFrame({
    "city": ["kathmandu", "pokhara", "dharan"],
    "temperature": [32, 45, 30],
}, index=[0, 1, 2])
temperature_df
```

```python
humidity_df = pd.DataFrame({
    "city": ["pokhara", "kathmandu", "dharan"],
    "humidity": [60, 80, 78]
}, index=[1, 0, 2])
humidity_df
```

```python
df = pd.concat([temperature_df, humidity_df], axis=1)
df
```
