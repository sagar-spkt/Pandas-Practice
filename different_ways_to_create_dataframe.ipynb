```python
weather_data1 = {
    'day': ['1/1/2017', '1/2/2017', '1/3/2017'],
    'temperature': [32, 35, 28],
    'windspeed': [6, 7, 2],
    'event': ['Rain', 'Sunny', 'Snow']
}
df_from_dictionary1 = pd.DataFrame(weather_data1)
df_from_dictionary1
```

```python
weather_data2 = [
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'windspeed': 7, 'temperature': 35, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},

]
df_from_dictionary2 = pd.DataFrame(data=weather_data2, columns=['day', 'temperature', 'windspeed', 'event'])
df_from_dictionary2
```
